```python
# Importing main libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## EX 1. According to the provided dataset, how many 2018 films were categorized as a Comedy? 

```python
# Reading the title_basics_2018.csv
df_tbasics = pd.read_csv('title_basics_2018.csv')
```

```python
# Getting the feel of the data.
df_tbasics.info()
```

The *title_basics_2018* dataset consists of 12,815 entries with (seemingly) no missing values

```python
# quick overview of the dataset and specifically genres column
df_tbasics.head(10)
```

From the *genres* column, it's clear that movies can have more than one genre so it's something to keep in mind when counting the occurrences of the *Comedy* movies. In the next section, we will count all movies with *Comedy* listed as **one of the genres or the only genre**. 

```python
# Counting all occurrences of 'Comedy', even when it co-occurs with other genres (Comedy-only genre or Comedy-multiple genre movies)
num_rows_comedy = df_tbasics[df_tbasics['genres'].str.contains('Comedy')].shape[0]
print("Number of rows containing 'Comedy':", num_rows_comedy)
```

So there is a total of 2233 movies with *Comedy* listed as **one of the genres or the only genre**. Let's see in how many of them, *Comedy* is <b>the only</b> genre and in how many it's <b>one of the multiple genres</b>

```python
# counting all cases of 'Comedy' occurring on its own or co-occuring with other genres. Let's take a look at first 10
df_tbasics['genres'][df_tbasics['genres'].str.contains('Comedy')].value_counts().head(10)
```

So out of 2233 all occurences of *Comedy*, 800 are single-genre *Comedy* movies and the rest, 2233-800 = 1433 are movies with *Comedy* listed as of the multiple generes. Let's confirm this in the script below.

```python
comedy_alone_count = 0
comedy_with_other_count = 0

# Count occurrences of 'Comedy'
for genres in df_tbasics['genres']:
    genre_list = genres.split(',')
    if 'Comedy' in genre_list:
        if len(genre_list) == 1:  # 'Comedy' appears alone
            comedy_alone_count += 1
        else:  # 'Comedy' appears with other genres
            comedy_with_other_count += 1

# Display the counts
print("Occurrences of 'Comedy' by itself:", comedy_alone_count)
print("Occurrences of 'Comedy' with other genres:", comedy_with_other_count)
```

Let's now look at co-occurrence of Comedy with other genres

```python
genre_counts = {}

#  occurrences of 'Comedy' with other genres
for genres in df_tbasics['genres']:
    genre_list = genres.split(',')
    if 'Comedy' in genre_list:
        for genre in genre_list:
            if genre != 'Comedy':
                if genre in genre_counts:
                    genre_counts[genre] += 1
                else:
                    genre_counts[genre] = 1

# Convert the dictionary to a DataFrame
df_genre_counts = pd.DataFrame({'Genre': list(genre_counts.keys()), 'Count': list(genre_counts.values())})

# Create a pivot table for visualization
pivot_table = df_genre_counts.pivot_table(index='Genre', aggfunc='sum').fillna(0)

# Sort the pivot table by counts in descending order
pivot_table = pivot_table.sort_values(by='Count', ascending=False)

# Plot heatmap
plt.figure(figsize=(4, 8))
sns.heatmap(pivot_table, annot=True, cmap='YlOrRd', fmt='.0f')
plt.title("Co-occurrences of 'Comedy' with Other Genres")
plt.xlabel('Genres')
plt.ylabel('Genres')
plt.show()
```

So in addition to single-genred 800 Comedies, there are 1433 movies when Comedy is listed as one of the multipe genres. Most of the times Comedy genre co-occurs with Drama and Romance. NOTE that the number counts in the heatmap above won't add up to 1433 since Comedy can co-occur with multiple other genres at once.

To sum up, there are 2233 movies that were categorized as *Comedy*. Of those 2233, 800 movies list *Comedy* as the only genre and in 1433 other *Comedy* co-occur with other genres, most often with Drama and Romance.

## EX 2. According to the provided dataset, how many 2018 films got a score of 8.0 or higher? 

```python
# reading title_ratings.csv
df_tratings = pd.read_csv('title_ratings.csv')
```

```python
# first look
df_tratings.info()
```

```python
# joining tables on tconst - unique identifier of the title. 
# Will be using the inner join because we need the rating 
merged_df = pd.merge(df_tbasics, df_tratings, on='tconst', how='inner')

# let's look at the resulting dataset
merged_df.head()
```

```python
# filtering the movies that were made in 2018 and have rating of 8 or higher
high_rating_2018 = merged_df[['primaryTitle', 'year', 'averageRating', 'numVotes', 'genres', 'runtimeMinutes']][(merged_df['averageRating'] >= 8) & (merged_df['year'] == 2018)]
```

```python
# sorting the result by the average rating from the highest to the lowest
high_rating_2018_sorted = high_rating_2018.sort_values(by='averageRating', ascending=False).reset_index(drop=True)
high_rating_2018_sorted.head()
```

```python
# counting  the movies that were made in 2018 having rating of 8 or higher
count_high_ratings_2018 = ((merged_df['averageRating'] >= 8) & (merged_df['year'] == 2018)).sum()
count_high_ratings_2018
```

Thus, the total number of movies that were released in 2018 and have the rating of 8 or higher is 780

## EX 3. What was the best film of 2018?

First, we need to set the definition and constraints for the best film. Is it the film that has the highest rating? Should there be a minimum for the number of votes for a movie to be in the consideration for the "best" one? Should the genre of the movie  be considered? Let's look at our dataset again.

Let's first pull up a list of the highest rating movies in 2018. We  already created it in the previous step.

```python
high_rating_2018_sorted.head(10)
```

We see that number of votes varies greatly. Let's take a closer look at the range of number of votes:

```python
high_rating_2018_sorted['numVotes'].describe().round(2)
```

So, the number of votes for movies in our segment (rating of 8 or great released in year 2018) ranges from 5 to 79 with Q1 (25th percentile) = 10, Q2 (50th percentile) = 23. We can establish minimum number of votes required for the best movie ~~nomination~~ consideration to be somewhere between Q1 and Q2. Let's say we establish it to be 20.

```python
high_rating_2018_minvotes = high_rating_2018_sorted[high_rating_2018_sorted['numVotes'] >= 20]
# let's sort by the AverageRating and numVotes
high_rating_2018_minvotes_sorted = \
high_rating_2018_minvotes.sort_values(by=['averageRating', 'numVotes'], ascending=[False, False]).reset_index(drop=True)
high_rating_2018_minvotes_sorted.head(5)
```

```python
# best movie across genres
high_rating_2018_minvotes_sorted.loc[0]
```

So, by our definition and constraints, the best movie in 2018 was the documentary *The Long Road to War* with a rating of 9.6 and 69 votes. However, it might be reasonable to establish 'best' movie per genre. We will consider single-genre movies only:

```python
# a datafame with single-genre movies
single_genres = high_rating_2018_minvotes_sorted[~high_rating_2018_minvotes_sorted['genres'].str.contains(',')]
```

A list of best movies by genre:

```python
single_genre = single_genres['genres'].unique().tolist()

first_rows = []  

for genre in single_genre:
    genre_df = single_genres[single_genres['genres'] == genre]  
    sorted_genre_df = genre_df.sort_values(by=['averageRating', 'numVotes'], ascending=[False, False])
    first_row = sorted_genre_df.head(1)
    first_rows.append(first_row)

# concatenating into a single DataFrame
best_by_genre = pd.concat(first_rows, ignore_index=True)

# dropping index and 'year'
best_by_genre.drop(columns=['index', 'year'], errors='ignore', inplace=True)

# list best movies by genre
best_by_genre
```
